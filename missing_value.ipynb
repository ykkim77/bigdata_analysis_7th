# 결측데이터 처리하기

## Iris 데이터 받기

```python
import pandas as pd    # pandas package import

# 데이터프레임의 컬럼명 지정

col = ["꽃받침길이", "꽃받침너비", "꽃잎길이", "꽃잎너비", "붓꽃종류"]

# 웹에 있는 CSV 파일 읽어와서 데이터프레임 df로 넣는다.

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", names = col)

df
```

- NULL 값을 가지는 데이터 추가 (데이터 값 None)

```python
import numpy as np    # numpy 패키지 import

new_data = [[4.8, None, 3.0, 1.9, "iris-setosa"]]   # 하나의 데이터 행 추가, 꽃받침의너비에 강제로 None값 추가
new_df = pd.DataFrame(new_data, columns= df.columns) # 새롭게 추가된 행으로 dataframe을 만듦
df = pd.concat([df, new_df], ignore_index=True) # 기존 df에 새롭게 만든 행을 추가   

df
```

- 기존의 첫번째 데이터의 꽃받침 길이와 꽃받침 너비의 값을 삭제(결측 데이터)

```python
df.iloc[0, 0] = np.NaN    # df의 0번째 행, 0번째 열의 데이터를 NaN으로 강제 교체 (NaN은 numpy에서 지원하는 NULL 데이터)
df.iloc[0, 1] = np.nan    # df의 0번째 행, 1번째 열의 데이터를 nan로 강제 교체 (nan은 pandas에서 표준 결측값으로 인식하는 NULL 데이터)
```

- 데이터에서 결측치가 잘 들어 갔는지 확인

```python
df
```

- 결측값 식별 - info() 메소드, isnull() 메소드

```python
df.info()     # RangeIndex: 151 <----원래 행은 151개의 행이나,
              # 꽃받침의길의 열에는 Non-Null이라는 행은 150개임, 즉 1개의 행은 결측치가 있다는 뜻임
              # 꽃받침의너비 열에는 Non-Null이라는 행은 149개임, 즉 2개의 행은 결측치가 있다는 뜻임
```

```python
df.isnull()  # 데이터에 Null값이 있으면 True, 없으면 False를 출력하는 함수
```

- 결측치의 통계를 사용하기

```python
df.isnull().sum()    # 각 열의 결측기 갯수를 파악할 수 있음
```

- 결측데이터 대체 -- 단순대치법 (완전분석법, 평균대치법)

- dropna() : 결측값이 있는 행 제거(완전분석법)
- drop([0,1]) : 행 삭제

```python
df_new = df.dropna(axis=0)   # axis가 0이면 행을 제거 axis가 1이면 행을 제거
df_new
```

```python
# "꽃받침너비" 열에서 결측값을 제거하고 출력
df_new = df["꽃받침너비"].dropna()
df_new
```

- fillna(): 결측값을 지정한 값으로 대체(평균대치법)

```python
df    # 기존데이터 다시 확인
```

```python
d_mean = df['꽃받침길이'].mean()    # 꽃받침길이의 평균값을 구함
df['꽃받침길이'].fillna(d_mean, inplace=True)  # d_mean의 값으로 결측값을 채움
df
```

- 결측값이 현재상태에서 얼마나 남았는지 확인

```python
df.isnull().sum()    # 꽃받침너비에 아직 2개의 행에 걸쳐 결측데이터가 남아 있음을 확인 가능
```
